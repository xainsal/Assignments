```python
import PIL
import os
import matplotlib.pyplot as plt
import numpy as np

def check_Dir(folder):
    x = os.path.exists(folder)
    if x == True:
        return True
    else:
        return False

def read_Dir_Files(folder):
    total_files = 0
    fileslist = []
    rootDir = folder
    for dirName, subdirList, fileList in os.walk(rootDir):
        for fname in fileList:
            fileslist.append(rootDir + fname)
            total_files += 1
    return fileslist, total_files


def to_Numpy_Array(name, no):
        imglist[no] = np.asarray(name, dtype=np.float32) / 255

def resize_Image(files_list, h, w):
    no = 0
    for file_name in files_list:
        img = PIL.Image.open(file_name)
        img = img.resize((h, w)).convert('RGB')
        #print(f'Name = {file_name} \nNumber {no+1}: Shape = {np.shape(img)}\n')
        to_Numpy_Array(img, no)
        no += 1
        
def show_All_Images(total_files, np_array, fileslist, len_Dir):
    fig = plt.figure(figsize=(12, 12))
    x = 0
    columns = 4
    rows = 5
    for i in range(1, columns*rows +1):
        a = fig.add_subplot(rows, columns, i)
        plt.imshow(np_array[i-1])
        a.set_title(fileslist[i-1][len_Dir:])
        x += 1

h = 200
w = 300
readDir = "photos/"
lenDir = len(readDir)

dirStatus = check_Dir(readDir)

if dirStatus == True:
    fileslist, total = read_Dir_Files(readDir)
    print(f'Total Images = {total}')
    imglist = np.zeros([total, h, w, 3])
    resize_Image(fileslist, w, h)
    show_All_Images(total, imglist, fileslist, lenDir)
else:
    print(f'"{readDir}" This Directory Does Not Exist')
    print('Please Check The Directory and Run Program Again')
```
